## 방탄소년단 뉴스 리스트

- title
- url
- description

```python
resp=requests.get("https://search.naver.com/search.naver?sm=tab_hty.top&where=news&query=%EB%B0%A9%ED%83%84%EC%86%8C%EB%85%84%EB%8B%A8&oquery=%EA%B2%80%EC%83%89&tqi=Uy424dprvh8ssiCHaWlssssssE0-222335") #위의 get과 같은 코드
resp
```

```python
soup=BeautifulSoup(resp.text)
ul_tags=soup.find('ul',class_='type01')
news_li=ul_tags.find_all('li',recursive=False)
#바로 하위에 있는것 가져오기. 이거 없으면 개수 27개임
print(len(news_li))
news_li
```

```python
result_list=[]
for news in news_li:
    title_tag=news.find('a',class_='_sp_each_url')
    title=title_tag.text
    url=title_tag.get('href')
    #print(title)
    #print(url)
    #break
    desc=news.find_all('dd')
    #print(desc[1])
    #break
    desc=desc[1].text
    result_list.append({'title':title,'url':url,'desc':desc})
```

```python
print(len(result_list))
result_list
```

### 두번째 방법

```python
for a_tag, desc_tag in zip(titles, descs):
    result_list.append({'titles':a_tag.text,'url':a_tag.get('href'),'desc':desc_tag.text})
```

## 블로그 페이지 가져오기

### 5페이지까지 출력

```python
result_list=[]#포문 안에 넣으면 초기화 반복돼서 마지막 열개만 리스트에 저장됨

for page in range(5):
    resp=requests.get("https://search.naver.com/search.naver?date_from=&date_option=0&date_to=&dup_remove=1&nso=&post_blogurl=&post_blogurl_without=&query=%EB%B6%80%EC%82%B0%20%EC%B9%A8%EC%88%98&sm=tab_pge&srchby=all&st=sim&where=post&start="+str(page*10+1))
    soup=BeautifulSoup(resp.text)
    
    ul_tags=soup.find('ul',class_='type01')
    blog_li=ul_tags.find_all('li',recursive=False)
    

    for blog in blog_li:
        title_tag=blog.find('a',class_='sh_blog_title')
        title=title_tag.text
        url=title_tag.get('href')

        desc=blog.find_all('dd')
        desc=desc[1].text
        result_list.append({'title':title,'url':url,'desc':desc})
```

```python
print(len(result_list))
result_list
```

### 함수

```python
def crawl_blog(query, pagenum=2):
    url='https://search.naver.com/search.naver'
    result_list1=[]
    
    for page in range(pagenum):
        params={
            'where':'post',
            'query':query,
            'start':page*10+1
        }
        resp=requests.get(url,params=params)#url 주소 뒤에 중복 고쳐야함

        soup=BeautifulSoup(resp.text)
    
        ul_tags=soup.find('ul',class_='type01')
        blog_li=ul_tags.find_all('li',recursive=False)
    

        for blog in blog_li:
            title_tag=blog.find('a',class_='sh_blog_title')
            title=title_tag.text
            url_=title_tag.get('href')

            desc=blog.find_all('dd')
            desc=desc[1].text
            result_list1.append({'title':title,'url':url_,'desc':desc})
        
    return result_list1
```

```python
blog_search_list=crawl_blog('트와이스',3)
print(len(blog_search_list))
blog_search_list
```

## 뉴스 상세페이지

```python
url='https://search.naver.com/search.naver'
params={'where':'news','query':'부산 침수'}
resp=requests.get(url,params=params)
resp
```

```python
soup=BeautifulSoup(resp.text)
a_tags=soup.select('ul.type01>li>dl>dt>a._sp_each_title')
len(a_tags)
```

```python
result_list=[]
for a_tag in a_tags:
    title=a_tag.text
    news_url=a_tag.get('href')
    
    detail_resp=requests.get(news_url)
    
    content=detail_resp.text
    result_list.append({'title':title,'url':news_url,'content':content})
```
