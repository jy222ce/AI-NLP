```python
th_tags=tr_tags[0].find_all('th')
title_list=[]
for th in th_tags:
    title_list.append(th.text)
print(title_list)
        
```

```python
result_list=[]

for tr_tag in tr_tags:
    td_tags=tr_tag.find_all('td')
    if len(td_tags)<7:
        continue
    
    data_list=[]
    for td in td_tags:
        data_list.append(td.text.strip()) #문자 스트립
        
    result_list.append(dict(zip(title_list, data_list)))
```

```python
# 미완코드

def get_day_sise(code, page_num):
    base_url="https://finance.naver.com/item/sise_day.nhn"
    
    for page in range(1,page_num+1):
        params={'code':code,'page':page}
    
        resp=requests.get(base_url,params=params)
        soup=BeautifulSoup(resp.text)
        
        tr_tags=soup.select('table.type2 tr')
        th_tags=tr_tags[0].find_all('th')
        
        title_list=[]
        for th in th_tags:
            title_list.append(th.text)
        
        result_list=[]

        for tr_tag in tr_tags:
            td_tags=tr_tag.find_all('td')
            if len(td_tags)<7:
                continue
        
            data_list=[]
            for td in td_tags:
                data_list.append(td.text.strip()) #문자 스트립
        
            result_list.append(dict(zip(title_list, data_list)))
    return result_list
```

```python
sise=get_day_sise('005930',2)
sise
```
