# 모델 설계하기
---------
- model.fit()으로 시작하는 부분은 모델을 실제로 수행하는 부분
- Dense(): 이 층에 몇 개의 노드를 만들 것인지를 결정
ex) model.add(Dense(8,input_dim=4,activation='relu')) #입력 노드 4개, 중간 히든 레이어의 노드가 8개
    model.add(Dense(1,activation='sigmoid')) #출력층
    
- 모델 실행:model.fit(X,Y,epochs=30,batch_size=10)

## 실습: XOR을 케라스로 해보기
- relu를 사용하면 정밀도가 높아짐
- 텐서로 레이어를 만들고 케라스로 사용

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation, Dense
```

```python
training_data=np.array([[0,0],[0,1],[1,0],[1,1]],'float32')
target_data=np.array([[0],[1],[1],[0]],'float32')

model=Sequential()
model.add(Dense(32,input_dim=2,activation='relu'))
model.add(Dense(1,activation='sigmoid'))

model.compile(loss='binary_crossentropy',optimizer='adam',
             metrics=['binary_accuracy'])
model.fit(training_data,target_data,nb_epoch=1000,verbose=2)
#nb_epoch가 오류날 경우 그냥 에포크로 하기

print(model.predict(training_data))
print('\nAccuraty: %.4f'%(model.evaluate(training_data,target_data)[1]))
```
