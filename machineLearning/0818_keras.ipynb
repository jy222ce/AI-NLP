## 실습: XOR을 케라스로 해보기
- relu를 사용하면 정밀도가 높아짐
- 텐서로 레이어를 만들고 케라스로 사용

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation, Dense
```

```python
training_data=np.array([[0,0],[0,1],[1,0],[1,1]],'float32')
target_data=np.array([[0],[1],[1],[0]],'float32')

model=Sequential()
model.add(Dense(32,input_dim=2,activation='relu'))
model.add(Dense(1,activation='sigmoid'))

model.compile(loss='binary_crossentropy',optimizer='adam',
             metrics=['binary_accuracy'])
model.fit(training_data,target_data,nb_epoch=1000,verbose=2)
#nb_epoch가 오류날 경우 그냥 에포크로 하기

print(model.predict(training_data))
print('\nAccuraty: %.4f'%(model.evaluate(training_data,target_data)[1]))
```

```python
#sigmoid 함수 정의
def sigmoid(z):
    return 1/(1+np.exp(-z))

#z=wx+b
def z(x,w,b):
    return w*x+b

#h(x)=sigmoid(z)=sigmoid(wx+b)
def H(x):
    global w,b
    return sigmoid(w*x+b)
```

## 오차제곱 평균

```python

w_list=np.linspace(-1.2,1.6,50)
x=np.linspace(-5,10,51)
y=np.append(np.zeros(30),np.ones(21))
b=0
Err=[]

for w_value in w_list:
    z=w_value*x+b
    Err.append(sum((sigmoid(z)-y)**2)/len(y))
               
print(Err)

plt.title('w와 Err의 그래프')
plt.xlabel('w')
plt.ylabel('Err')

plt.plot(w_list,Err)

plt.show()
```

```python
w_list=np.linspace(-7,3,101)
b=0
x=np.linspace(-5,10,51)
y=np.append(np.zeros(30),np.ones(21))

cost=[]
for w_value in w_list:
    z=w_value*x+b
    h=sigmoid(z)
    c=-y*np.log(h)-(1-y)*np.log(1-h)
    cost.append(sum(c)/len(c))
    
plt.title('기울기 w에 대한 cost 그래프')
plt.grid(True)
plt.xlabel('w')
plt.ylabel('cost')
plt.scatter(w_list,cost)
plt.show()
```
