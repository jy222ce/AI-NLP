## 함수화

```python
def get_news_list(sid1,sid2=None,date=None):
    base_url='https://news.naver.com/main/list.nhn'
    params={
        'mode':'LS2D',
        'mid':'shm',
        'sid1':sid1,
        'sid2':sid2,
        'date':date,
        'page':0
    } 
    
    is_ended=False
    
    while not is_ended:
        params['page']+=1
        
        resp=requests.get(base_url,params=params)
        soup=BeautifulSoup(resp.text,'lxml')
        
        news_list_tags=soup.select('div.list_body>ul>li')

        num_dup=0 #중복 개수
        
        for news_item in news_list_tags:
            dt_tags=news_item.find_all('dt')
            if len(dt_tags)==1:#이미지없음
                title_tag=dt_tag[0]
            else:#이미지있음
                title_tag=dt_tags[1]
            
            news={'title':title_tag.text.strip(),
                                'url':title_tag.find('a').get('href')}
            
            if num_dup>=3: 
                #실시간으로 크롤링할 때 뉴스가 업데이트 될 때를 대비해서 세개 이상일 경우 같은 기사라고 여김
                is_ended=True
                break
            
            if news in result_list:
                num_dup+=1
                continue
            
            result_list.append(news)
            
            
        if len((news_list_tags))<20:
            is_ended=True

    return result_list
```

```python
get_news_list(105,731,'20200803')
```
